# 1. Importing Libraries and Preparing a sample data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

creating sample data for 2 columns using random values:
*   experience
*   salary

```python
exp_list = np.random.uniform(0,6,33)
exp_list = list(np.around(exp_list,decimals=1))

salary_list = np.random.uniform(20000,130000,33)
salary_list = list(np.around(salary_list,decimals=0))
salary_list = list(map(int,salary_list))
```

creating dataframe

```python
df = pd.DataFrame({'experience':exp_list,
                   'salary':salary_list})
```

```python
df.head()
```

plotting datapoints

```python
plt.scatter(df['experience'],df['salary'])
```

```python
df.info()
```

```python
df.describe()
```

# Splitting the sample data

importing train test split from skearn's model_selection module

```python
from sklearn.model_selection import train_test_split
```

```python
x = np.array(df['experience'])
y = np.array(df['salary'])
```

```python
x = x.reshape(-1,1)
y = y.reshape(-1,1)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.2,random_state=3)
```

plotting a scatter plot for train and test value of X and Y

```python
plt.scatter(X_train,Y_train)
plt.xlabel('X')
plt.ylabel('Y')
plt.scatter(X_test,Y_test)
plt.legend(['train','test'])
```

importing linear regressor

```python
from sklearn.linear_model import LinearRegression
```

# Training the model

```python
# X_train=np.array(X_train).reshape(1,-1)
# Y_train=np.array(Y_train).reshape(1,-1)
# X_train
```

```python
model=LinearRegression()
model.fit(X_train,Y_train)
```

# Testing the model

```python
Y_test
```

```python
y_pred = model.predict(X_test)
```

# Checking the Model Accuracy

```python
from sklearn.metrics import mean_squared_error as MSE
```

```python
MSE(Y_test,y_pred)
```

```python
from sklearn.metrics import r2_score as r2
```

r2 score:

```python
r2(Y_test,y_pred)
```

# Conclusion

As we can see, the r2 score in negative which implies our model is not following the tends of the data.

In this case, the data doest follow a linear trend i.e the r2 score is negative, impling the incorrection in model selection.

It can also be a case of overfitting if the model give nearly accurate result for the data in training phase.
